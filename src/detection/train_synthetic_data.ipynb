```python
import yaml
import os
from xml2yolo import json_to_yolo

```

```python
# Read YAML file
with open("configDNVdata.yaml", 'r') as stream:
    config_file = yaml.safe_load(stream)

DATASET_DIR = config_file['dataset_src_dir']
    
TRAIN_PATH = os.path.join(DATASET_DIR, config_file['train'])
VAL_PATH = os.path.join(DATASET_DIR, config_file['val'])
```

```python
json_to_yolo(input_dir, label_dir, image_height, image_width)
```

```python
from xml2yolo import json_to_yolo
import os


DATASET_DIR_MORNING_cloudy = '/cluster/home/solveijm/DNV_data/2023-04-26_1658_dnv_scenario1_full_00_000_MorningCloudy'

input_dir = os.path.join(DATASET_DIR_MORNING_cloudy, 'images/')
label_dir = os.path.join(DATASET_DIR_MORNING_cloudy , 'labels/')
image_dir2 = os.path.join(DATASET_DIR_MORNING_cloudy, 'images/')


image_height = 2056
image_width = 2464
```

```python
json_to_yolo(input_dir, label_dir, image_height, image_width)
```

```python
import glob
import random
import itertools

files = glob.glob(os.path.join(image_dir1, '*.jpg'))
files2 =  glob.glob(os.path.join(image_dir2, '*.jpg'))

all_files = files + files2
random.shuffle(all_files)

train_split = 0.8
```

```python
all_files
```

```python
'./'+'/'.join(all_files[0].split('/')[5:])
```

```python
split_point_train = round(len(all_files)*train_split)


with open(TRAIN_PATH, 'w+') as file:
    file.write('\n'.join(str('./'+'/'.join(line.split('/')[5:])) for line in all_files[:split_point_train]))

with open(VAL_PATH, 'w+') as file:
    file.write('\n'.join(str('./'+'/'.join(line.split('/')[5:])) for line in all_files[split_point_train:]))
```

```python
print('# of train: ', split_point_train)
print('# of val: ', str(len(all_files)-split_point_train))
```
