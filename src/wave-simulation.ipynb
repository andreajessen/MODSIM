```python
import numpy as np
import matplotlib.pyplot as plt

# set parameters
phi = 0.99
b = 0.99
sigma = 1

# Set the initial value
X_0 = 0
```

```python
# Generate the time series
X = [X_0]
for i in range(1, 10000):
    X.append(phi*X[i-1] + np.random.normal(scale=sigma))

# Plot the time series
plt.plot(X)
plt.show()


Y = [X_0]
for i in range(1, 10000):
    Y.append(b*Y[i-1] + X[i])

# Plot the time series
plt.plot(Y)
plt.show()
```

```python
# Variance
variance = (sigma**2) / (1-phi**2)
variance
```
